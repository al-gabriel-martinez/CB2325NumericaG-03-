```python
import sys
import matplotlib.pyplot as plt
import numpy as np
sys.path.append('..')
from CB2325NumericaG3.interpolacao import linear_interp, poly_interp, hermite_interp
```

# Interpolação Linear por Partes

```python
#Criando Interpolação
X = [1,2,3,4,5,6]
Y = [3,6,5,1,0,-4]
f = linear_interp(X, Y, plot="True", title="Interpolação Linear por Partes")

# Avaliando a Função Interpoladora
t = 3.5
print(f"f({t}) = {f(t)}")
```

```python
#Criando Interpolação 2
X = [5,3,7,6,10,-1,0]
Y = [3,4,2,8,5,1,5]
f2 = linear_interp(X, Y, plot="True", title="Interpolação Linear por Partes 2")

# Avaliando a Função Interpoladora
t = 4
print(f"f2({t}) = {f2(t)}")
```

```python
#Operando com f e f2
Soma = lambda t: f(t) + f2(t)

# Plotando Gráfico da Soma
x = [x / 100 for x in range(100*-1//1 -1, 100*10//1 + 1)]
y = [Soma(t) for t in x]

plt.plot(x, y)

plt.title("Gráfico da Soma")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.grid(True)

plt.show()
```

```python
#Interpolação da função seno
N = 30
X = np.linspace(0, 2*np.pi, N)
Y = np.sin(X)

sen = linear_interp(X, Y, plot = "True", title = "Função Seno")

print("O seno de 30° vale(aprox):", sen(np.pi/6))
```

# Interpolação Polinomial

```python
# Exemplo de interpolação com 4 pontos
X = [1, 2, 3, 4]
Y = [0, 6, 24, 60]

pol = poly_interp(X, Y, method = "lagrange", title = "Interpolação Polinomial")

# Avaliando o Polinômio
t = 5
print(f"pol({t}) = {pol(t)}")
```

```python
# Exemplo de interpolação com os 3 métodos.

X = [-1, 0, 1, 2, 4, 6]
Y = [3, 5, 1, 0, 4, 5]

P_Lg = poly_interp(X, Y, method = "lagrange", plot = True, title = "Interpolação Lagrange")
P_Nw = poly_interp(X, Y, method = "newton", plot = True, title = "Interpolação Newton")
P_Vm = poly_interp(X, Y, method = "vandermonde", plot = True, title = "Interpolação Vandermonde")

t = 5
print(f"P_Lg({t}) = {P_Lg(t)}")
print(f"P_Nw({t}) = {P_Nw(t)}")
print(f"P_Vm({t}) = {P_Vm(t)}")
```

```python
# Comparando a velocidade dos métodos
import time

# Pontos aleatórios
N = 500
X = np.random.uniform(0, 100, N)
Y = np.random.uniform(0, 100, N)

# Tempo por Lagrange
t1 = time.time()
P_Lg = poly_interp(X, Y, method = "lagrange", plot = False)
t2 = time.time()
time_Lagrange = t2 - t1

# Tempo por Newton
t1 = time.time()
P_Nw = poly_interp(X, Y, method = "newton", plot = False)
t2 = time.time()
time_Newton = t2 - t1

# Tempo por Vandermonde
t1 = time.time()
P_Vm = poly_interp(X, Y, method = "vandermonde", plot = False)
t2 = time.time()
time_Vandermonde = t2 - t1

print("O tempo do método de Lagrange foi:", time_Lagrange)
print("O tempo do método de Newton foi:", time_Newton)
print("O tempo do método de Vandermonde foi:", time_Vandermonde)
```

# Interpolação de Hermite

```python
# Criando a Interpolação
X = [0, 1, 2, 3, 4]
Y = [[0], [1], [4], [9], [16]]  # f(x)
f = hermite_interp(X, Y, plot=False)

print(f(2.5))
print(f(5))
```

```python
# Plotando o gráfico de f
h_func = hermite_interp(X, Y, title="Interpolação Polinomial de Hermite")
```

```python
# Criando Interpolação com derivadas
X = [-2, -1, 0, 1, 2]
# f(x), f'(x)
derivadas = [[7, -5], [3, -3], [1, -1], [1, 1], [3, 3]]
f2 = hermite_interp(X, derivadas, plot=False)

print(f2(3))
print(f2(-3))
```

```python
# Plotando gráfico de f2
h_func2 = hermite_interp(X, derivadas, title="Interpolação Polinomial de Hermite 2")
```

```python
# Interpolação com número de derivadas por ponto variável
X = [-1, 0, 2, 5]
# f(x), f'(x), f''(x),...
derivadas = [[0, 1 ,1], [1], [1, 3], [-1, 2, 1, 3]]
f3 = hermite_interp(X, derivadas, plot=False)

print(f3(1))
```

```python
# Plotando gráfico de f3
h_func3 = hermite_interp(X, derivadas, title="Interpolação Polinomial de Hermite 3")
```

```python
# Interpolação de cos²(x)
X = np.linspace(0, 2*np.pi, 5)
D = [[np.cos(x)**2, -np.sin(2*x), -2*np.cos(2*x)] for x in X]

cos = hermite_interp(X, D, plot=False)

print("cos²(π/4):", cos(np.pi/4))
```
